# Running Java from Python

```python
import h5py
import subprocess




with h5py.File('../public-queries-10k-clip768v2.h5', 'r') as f:
    # Read dataset from h5 file
    data_vectors_orig = f['emb'][()]
with h5py.File('../public-queries-10k-pca96v2.h5', 'r') as f:
    # Read dataset from h5 file
    data_vectors_pca96 = f['pca96'][()]

# Print the number of rows of the dataset
print(f'Number of rows: {data_vectors_orig.shape[0]}')


# Create a sequence of integers from 1 to the number of rows of the dataset
# for i in range(data_vectors_orig.shape[0]):
for i in range(1):
    object_id = i + 1
    object_vector_orig = ','.join([str(x) for x in data_vectors_orig[i]])
    object_vector_pca96 = ','.join([str(x) for x in data_vectors_pca96[i]])
    print("{" + str(object_id) + ";" + object_vector_orig + ";" + object_vector_pca96 + "}")
    #print(repr('{') + {object_id} + {object_vector_pca96} + repr('}'))
# {id od 1, vector_str_orig, vector_str_pca96}









result_file_path = 'test.h5'

java_output = subprocess.check_output(['java', 'TestJavaClass', '5', '3'], universal_newlines=True)

# Convert output to an integer
result = int(java_output.strip())

print(f"The result from Java: {result}")

# Store the result into a h5 file
with h5py.File(result_file_path, 'w') as f:
    f.create_dataset('result', data=result)
    # Store the attribute into a h5 file
    f.attrs['data'] = 'pca96'

# Read and print the result from the h5 file
with h5py.File(result_file_path, 'r') as f:
    result = f['result'][()]
    print(f"The result from h5: {result}")


```
